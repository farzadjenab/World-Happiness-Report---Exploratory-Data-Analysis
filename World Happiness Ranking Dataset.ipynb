```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options for better viewing
pd.set_option('display.max_columns', None)

# Set plot style
plt.style.use('seaborn-v0_8-whitegrid')

# For showing plots in jupyter
%matplotlib inline
```

```python
# Load the dataset
df = pd.read_csv('world_happiness_report.csv')

# Display first 5 rows to see what we have
df.head()
```

```python
# Check the shape of dataset (rows, columns)
print("Dataset Shape:")
print(f"Rows: {df.shape[0]}")
print(f"Columns: {df.shape[1]}")
```

```python
# Check column names
print("Column Names:")
print(df.columns.tolist())
```

```python
# Check data types of each column
print("Data Types:")
print(df.dtypes)
```

```python
# Get basic info about the dataset
df.info()
```

```python
# Get statistical summary of numerical columns
df.describe()
```

```python
# Round the statistics for better readability
df.describe().round(2)
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()
print("Missing Values per Column:")
print(missing_values)
```

```python
# Calculate percentage of missing values
missing_percentage = (df.isnull().sum() / len(df) * 100).round(2)
print("\nMissing Values Percentage:")
print(missing_percentage)
```

```python
# Visualize missing values
plt.figure(figsize=(10, 5))
df.isnull().sum().plot(kind='bar', color='coral')
plt.title('Missing Values Count per Column')
plt.xlabel('Columns')
plt.ylabel('Count of Missing Values')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Check unique countries
print(f"Number of Unique Countries: {df['Country'].nunique()}")
print("\nFirst 20 Countries:")
print(df['Country'].unique()[:20])
```

```python
# Check unique regions
print(f"Number of Unique Regions: {df['Region'].nunique()}")
print("\nAll Regions:")
print(df['Region'].unique())
```

```python
# Check unique years
print(f"Years in Dataset: {sorted(df['year'].unique())}")
```

```python
# Count of countries per region
region_counts = df['Region'].value_counts()
print("Countries per Region:")
print(region_counts)
```

```python
# Visualize region distribution
plt.figure(figsize=(12, 6))
df['Region'].value_counts().plot(kind='bar', color='skyblue', edgecolor='navy')
plt.title('Number of Records per Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Histogram of Happiness Score
plt.figure(figsize=(10, 5))
plt.hist(df['Happiness Score'], bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('Distribution of Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.axvline(df['Happiness Score'].mean(), color='red', linestyle='--', label=f'Mean: {df["Happiness Score"].mean():.2f}')
plt.legend()
plt.show()
```

```python
# Top 10 Happiest Countries (latest year in dataset)
latest_year = df['year'].max()
df_latest = df[df['year'] == latest_year]

top_10_happy = df_latest.nsmallest(10, 'Happiness Rank')[['Country', 'Region', 'Happiness Score', 'Happiness Rank']]
print(f"Top 10 Happiest Countries in {latest_year}:")
print(top_10_happy)
```

```python
# Bottom 10 (Least Happy) Countries
bottom_10_happy = df_latest.nlargest(10, 'Happiness Rank')[['Country', 'Region', 'Happiness Score', 'Happiness Rank']]
print(f"\nBottom 10 Least Happy Countries in {latest_year}:")
print(bottom_10_happy)
```

```python
# Average Happiness Score by Region
region_happiness = df.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False)
print("Average Happiness Score by Region:")
print(region_happiness.round(3))
```

```python
# Visualize Region Happiness
plt.figure(figsize=(12, 6))
region_happiness.plot(kind='bar', color='teal', edgecolor='black')
plt.title('Average Happiness Score by Region')
plt.xlabel('Region')
plt.ylabel('Average Happiness Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Box plot of Happiness Score by Region
plt.figure(figsize=(14, 6))
df.boxplot(column='Happiness Score', by='Region', figsize=(14, 6))
plt.title('Happiness Score Distribution by Region')
plt.suptitle('')  # Remove automatic title
plt.xlabel('Region')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Select numerical columns for correlation
numerical_cols = ['Happiness Score', 'Economy (GDP per Capita)', 'Family', 
                  'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 
                  'Generosity']

# Calculate correlation matrix
correlation_matrix = df[numerical_cols].corr()
print("Correlation Matrix:")
print(correlation_matrix.round(2))
```

```python
# Visualize correlation with heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# Which factors correlate most with Happiness Score?
happiness_corr = correlation_matrix['Happiness Score'].sort_values(ascending=False)
print("\nCorrelation with Happiness Score:")
print(happiness_corr)
```

```python
# Scatter plot: GDP vs Happiness Score
plt.figure(figsize=(10, 6))
plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'], 
            alpha=0.5, color='purple', edgecolor='black')
plt.title('GDP per Capita vs Happiness Score')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot with region colors
plt.figure(figsize=(12, 7))
regions = df['Region'].unique()
colors = plt.cm.tab10(np.linspace(0, 1, len(regions)))

for region, color in zip(regions, colors):
    region_data = df[df['Region'] == region]
    plt.scatter(region_data['Economy (GDP per Capita)'], 
                region_data['Happiness Score'], 
                alpha=0.6, label=region, color=color)

plt.title('GDP per Capita vs Happiness Score (by Region)')
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8)
plt.tight_layout()
plt.show()
```

```python
# Average Happiness Score per Year
yearly_happiness = df.groupby('year')['Happiness Score'].mean()
print("Average Happiness Score per Year:")
print(yearly_happiness.round(3))
```

```python
# Line plot of yearly trend
plt.figure(figsize=(10, 5))
yearly_happiness.plot(kind='line', marker='o', color='blue', linewidth=2, markersize=8)
plt.title('Average Global Happiness Score Over Years')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Happiness trend by region over years
plt.figure(figsize=(14, 7))
region_year_happiness = df.groupby(['year', 'Region'])['Happiness Score'].mean().unstack()
region_year_happiness.plot(marker='o', figsize=(14, 7))
plt.title('Happiness Score Trend by Region Over Years')
plt.xlabel('Year')
plt.ylabel('Average Happiness Score')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Summary of findings
print("="*60)
print("SUMMARY OF KEY FINDINGS")
print("="*60)

print(f"\n1. Dataset contains {df.shape[0]} records and {df.shape[1]} columns")
print(f"\n2. Number of unique countries: {df['Country'].nunique()}")
print(f"\n3. Number of regions: {df['Region'].nunique()}")
print(f"\n4. Years covered: {sorted(df['year'].unique())}")

print(f"\n5. Happiness Score Statistics:")
print(f"   - Mean: {df['Happiness Score'].mean():.2f}")
print(f"   - Min: {df['Happiness Score'].min():.2f}")
print(f"   - Max: {df['Happiness Score'].max():.2f}")

print(f"\n6. Happiest Region: {region_happiness.idxmax()} ({region_happiness.max():.2f})")
print(f"\n7. Least Happy Region: {region_happiness.idxmin()} ({region_happiness.min():.2f})")

print(f"\n8. Top 3 factors correlated with Happiness:")
for i, (factor, corr) in enumerate(happiness_corr[1:4].items(), 1):
    print(f"   {i}. {factor}: {corr:.2f}")

print("\n" + "="*60)
```
